```python
# Membrane Components
C_mem=bgt.new("C_mem", name="C_m", value=C_m)
zero_mem=bgt.new("0")
TF_zF_5=bgt.new("TF", value=zF_5, name="zF_5")
TF_zF_8=bgt.new("TF", value=zF_8, name="zF_8")

# From C:P1 to C:P15
C_P1=bgt.new("Ce", name="P1", library="BioChem", value={'k':K1 , 'R':R, 'T':T})
C_P2=bgt.new("Ce", name="P2", library="BioChem", value={'k':K2 , 'R':R, 'T':T})
C_P3=bgt.new("Ce", name="P3", library="BioChem", value={'k':K3 , 'R':R, 'T':T})
C_P4=bgt.new("Ce", name="P4", library="BioChem", value={'k':K4 , 'R':R, 'T':T})
C_P5=bgt.new("Ce", name="P5", library="BioChem", value={'k':K5 , 'R':R, 'T':T})
C_P6=bgt.new("Ce", name="P6", library="BioChem", value={'k':K6 , 'R':R, 'T':T})
C_P7=bgt.new("Ce", name="P7", library="BioChem", value={'k':K7 , 'R':R, 'T':T})
C_P8=bgt.new("Ce", name="P8", library="BioChem", value={'k':K8 , 'R':R, 'T':T})
C_P9=bgt.new("Ce", name="P9", library="BioChem", value={'k':K9 , 'R':R, 'T':T})
C_P10=bgt.new("Ce", name="P10", library="BioChem", value={'k':K10 , 'R':R, 'T':T})
C_P11=bgt.new("Ce", name="P11", library="BioChem", value={'k':K11 , 'R':R, 'T':T})
C_P12=bgt.new("Ce", name="P12", library="BioChem", value={'k':K12 , 'R':R, 'T':T})
C_P13=bgt.new("Ce", name="P13", library="BioChem", value={'k':K13 , 'R':R, 'T':T})
C_P14=bgt.new("Ce", name="P14", library="BioChem", value={'k':K14 , 'R':R, 'T':T})
C_P15=bgt.new("Ce", name="P15", library="BioChem", value={'k':K15 , 'R':R, 'T':T})

# "Ki, Ke, Nai, Nae, Pi, H, MgADP, MgATP" Components
C_Ki = bgt.new("Ce_S", name="Ki", library="BioChem", value={'k':K_Ki , 'R':R, 'T':T})
C_Ke = bgt.new("Ce_S", name="Ke", library="BioChem", value={'k':K_Ke , 'R':R, 'T':T})
C_Nai = bgt.new("Ce_S", name="Nai", library="BioChem", value={'k':K_Nai , 'R':R, 'T':T})
C_Nae = bgt.new("Ce_S", name="Nae", library="BioChem", value={'k':K_Nae , 'R':R, 'T':T})
C_MgATP = bgt.new("Ce_S", name="MgATP", library="BioChem", value={'k':K_MgATP , 'R':R, 'T':T})
C_MgADP = bgt.new("Ce_S", name="MgADP", library="BioChem", value={'k':K_MgADP , 'R':R, 'T':T})
C_Pi = bgt.new("Ce_S", name="Pi", library="BioChem", value={'k':K_Pi , 'R':R, 'T':T})
C_H = bgt.new("Ce_S", name="H", library="BioChem", value={'k':K_H , 'R':R, 'T':T})

# Reactions
Re_R1 = bgt.new("Re", library="BioChem", value={'r':r1, 'R':R, 'T':T, 'name':'Re_R1'})
Re_R2 = bgt.new("Re", library="BioChem", value={'r':r2, 'R':R, 'T':T, 'name':'Re_R2'})
Re_R3 = bgt.new("Re", library="BioChem", value={'r':r3, 'R':R, 'T':T, 'name':'Re_R3'})
Re_R4 = bgt.new("Re", library="BioChem", value={'r':r4, 'R':R, 'T':T, 'name':'Re_R4'})
Re_R5 = bgt.new("Re", library="BioChem", value={'r':r5, 'R':R, 'T':T, 'name':'Re_R5'})
Re_R6 = bgt.new("Re", library="BioChem", value={'r':r6, 'R':R, 'T':T, 'name':'Re_R6'})
Re_R7 = bgt.new("Re", library="BioChem", value={'r':r7, 'R':R, 'T':T, 'name':'Re_R7'})
Re_R8 = bgt.new("Re", library="BioChem", value={'r':r8, 'R':R, 'T':T, 'name':'Re_R8'})
Re_R9 = bgt.new("Re", library="BioChem", value={'r':r9, 'R':R, 'T':T, 'name':'Re_R9'})
Re_R10 = bgt.new("Re", library="BioChem", value={'r':r10, 'R':R, 'T':T, 'name':'Re_R10'})
Re_R11 = bgt.new("Re", library="BioChem", value={'r':r11, 'R':R, 'T':T, 'name':'Re_R11'})
Re_R12 = bgt.new("Re", library="BioChem", value={'r':r12, 'R':R, 'T':T, 'name':'Re_R12'})
Re_R13 = bgt.new("Re", library="BioChem", value={'r':r13, 'R':R, 'T':T, 'name':'Re_R13'})
Re_R14 = bgt.new("Re", library="BioChem", value={'r':r14, 'R':R, 'T':T, 'name':'Re_R14'})
Re_R15 = bgt.new("Re", library="BioChem", value={'r':r15, 'R':R, 'T':T, 'name':'Re_R15'})

# 0-junctions & 1-junctions for the concentrations and states
zero_P1 = bgt.new("0"); zero_P2 = bgt.new("0"); zero_P3 = bgt.new("0"); zero_P4 = bgt.new("0"); zero_P5 = bgt.new("0");
zero_P6 = bgt.new("0"); zero_P7 = bgt.new("0"); zero_P8 = bgt.new("0"); zero_P9 = bgt.new("0"); zero_P10 = bgt.new("0");
zero_P11 = bgt.new("0"); zero_P12 = bgt.new("0"); zero_P13 = bgt.new("0"); zero_P14 = bgt.new("0"); zero_P15 = bgt.new("0")

zero_Ki = bgt.new("0"); zero_Ke = bgt.new("0"); zero_Nai = bgt.new("0"); zero_Nae = bgt.new("0");

one_MgATP=bgt.new("1"); one_MgADP=bgt.new("1"); one_Pi_H=bgt.new("1");

# 1-junctions for the Reactions
one_R1_reverse=bgt.new("1"); one_R2_reverse=bgt.new("1"); one_R3_forward=bgt.new("1"); one_R4_forward=bgt.new("1"); 
one_R5_forward=bgt.new("1"); one_R5_reverse=bgt.new("1"); one_R8_reverse=bgt.new("1"); one_R9_reverse=bgt.new("1");
one_R10_reverse=bgt.new("1"); one_R11_forward=bgt.new("1"); one_R12_forward=bgt.new("1"); 

# Add All to the model
bgt.add(model,C_mem,zero_mem,TF_zF_5,TF_zF_8,C_P1,C_P2,C_P3,C_P4,C_P5,C_P6,C_P7,C_P8,C_P9,C_P10,C_P11,C_P12,C_P13,C_P14,
        C_P15,C_Ki,C_Ke,C_Nai,C_Nae,C_MgATP,C_MgADP,C_H,C_Pi,Re_R1,Re_R2,Re_R3,Re_R4,Re_R5,Re_R6,Re_R7,Re_R8,Re_R9,
        Re_R10,Re_R11,Re_R12,Re_R13,Re_R14,Re_R15,zero_P1,zero_P2,zero_P3,zero_P4,zero_P5,zero_P6,zero_P7,zero_P8,
        zero_P9,zero_P10,zero_P11,zero_P12,zero_P13,zero_P14,zero_P15,zero_Ki,zero_Ke,zero_Nai,zero_Nae,one_MgATP,
        one_MgADP,one_Pi_H,one_R1_reverse,one_R2_reverse,one_R3_forward,one_R4_forward,one_R5_forward,one_R5_reverse,
        one_R8_reverse,one_R9_reverse,one_R10_reverse,one_R11_forward,one_R12_forward)
```

```python
# Connections from C:P1 to C:P15
bgt.connect(zero_P1,C_P1); bgt.connect(zero_P1,Re_R1); bgt.connect(Re_R1,one_R1_reverse); 
bgt.connect(one_R1_reverse,zero_Ki); bgt.connect(zero_Ki,C_Ki); bgt.connect(one_R1_reverse,zero_P2);
bgt.connect(zero_P2,C_P2); bgt.connect(zero_P2,Re_R2); bgt.connect(Re_R2,one_R2_reverse); 
bgt.connect(one_R2_reverse,zero_Ki); bgt.connect(one_R2_reverse,zero_P3); bgt.connect(zero_P3,C_P3); 
bgt.connect(zero_P3,one_R3_forward); bgt.connect(one_R3_forward,Re_R3); bgt.connect(Re_R3,zero_P4); 
bgt.connect(zero_P4,C_P4); bgt.connect(zero_P4,one_R4_forward); bgt.connect(one_R4_forward,Re_R4); 
bgt.connect(zero_Nai,C_Nai); bgt.connect(zero_Nai,one_R3_forward); bgt.connect(zero_Nai,one_R4_forward); 
bgt.connect(zero_Nai,one_R5_forward); bgt.connect(Re_R4,zero_P5); bgt.connect(zero_P5,C_P5); 
bgt.connect(zero_P5,one_R5_forward); 
bgt.connect(one_R5_forward,Re_R5); bgt.connect(Re_R5,one_R5_reverse); bgt.connect(one_R5_reverse,zero_P6); 
bgt.connect(zero_P6,C_P6); bgt.connect(zero_P6,Re_R6); bgt.connect(one_R5_reverse,(TF_zF_5,0)); 
bgt.connect((TF_zF_5,1),zero_mem); bgt.connect(zero_mem,C_mem); bgt.connect(Re_R6,one_MgADP); 
bgt.connect(one_MgADP,C_MgADP); bgt.connect(one_MgADP,zero_P7); bgt.connect(zero_P7,C_P7); 
bgt.connect(zero_P7,Re_R7); bgt.connect(Re_R7,zero_P8); bgt.connect(zero_P8,C_P8); 
bgt.connect(zero_P8,Re_R8); bgt.connect(Re_R8,one_R8_reverse); bgt.connect(one_R8_reverse,(TF_zF_8,0)); 
bgt.connect((TF_zF_8,1),zero_mem); bgt.connect(one_R8_reverse,zero_P9); bgt.connect(one_R8_reverse,zero_Nae); 
bgt.connect(zero_P9,C_P9); bgt.connect(zero_P9,Re_R9); bgt.connect(Re_R9,one_R9_reverse); 
bgt.connect(one_R9_reverse,zero_Nae); bgt.connect(one_R9_reverse,zero_P10); bgt.connect(zero_P10,C_P10); 
bgt.connect(zero_P10,Re_R10); bgt.connect(Re_R10,one_R10_reverse); bgt.connect(one_R10_reverse,zero_Nae); 
bgt.connect(zero_Nae,C_Nae); bgt.connect(one_R10_reverse,zero_P11); bgt.connect(zero_P11,C_P11); 
bgt.connect(zero_P11,one_R11_forward); bgt.connect(one_R11_forward,Re_R11); bgt.connect(Re_R11,zero_P12); 
bgt.connect(zero_P12,C_P12); bgt.connect(zero_P12,one_R12_forward); bgt.connect(one_R12_forward,Re_R12); 
bgt.connect(zero_Ke,one_R12_forward); bgt.connect(zero_Ke,C_Ke); bgt.connect(zero_Ke,one_R11_forward); 
bgt.connect(Re_R12,zero_P13); bgt.connect(zero_P13,C_P13); bgt.connect(zero_P13,Re_R13); 
bgt.connect(Re_R13,one_Pi_H); bgt.connect(one_Pi_H,C_Pi); bgt.connect(one_Pi_H,C_H); 
bgt.connect(one_Pi_H,zero_P14); bgt.connect(zero_P14,C_P14); bgt.connect(zero_P14,one_MgATP); 
bgt.connect(C_MgATP,one_MgATP); bgt.connect(one_MgATP,Re_R14); bgt.connect(Re_R14,zero_P15); 
bgt.connect(zero_P15,C_P15); bgt.connect(zero_P15,Re_R15); bgt.connect(Re_R15,zero_P1)
```
